```python
import pandas as pd
df = pd.read_csv("titanic-passengers.csv", sep= ';')
df.head()
```

```python
df.drop(["Ticket","Cabin"], axis= 1,inplace = True)
df.head()
```

```python
df["Age"].fillna(df["Age"].mean(),inplace = True)
print(df["Embarked"].value_counts())
df["Embarked"] .fillna('S',inplace = True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
df["Embarked"]=LabelEncoder().fit_transform(df["Embarked"])
df["Survived"]=LabelEncoder().fit_transform(df["Survived"])
df["Sex"]=LabelEncoder().fit_transform(df["Sex"])

#Survived : 0 = No , 1 = Yes
#Sex : 0 ; Female , 1 = Male

df.head()
```

```python
import matplotlib.pyplot as plt

def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
plot_correlation_map(df)

```

the paragraph!!
this graph describes the correlation between each column and the others...
as the +ve value in the square tells that the 2 parameters are going in the same direction together(either increase or decrease but with different factor)
and the -ve value tells that the 2 parameters go in a different direction (one increases and the other decreases)
the value 1 means that they have the same slop(they increase or decrease with the same factor)


In this graph, we can see that as the age increases, the Fare increases, the correlation between them is +ve, and so on.
but we cant indicate the relation between Sex and Fare (for instance) because Sex is actually a Categorical Value (even tho we encoded it into Numerical values!)
